```python
import numpy as np
import pandas as pd

def log2_fold_change(matrix1, matrix2):
    """
    Compute log2 fold change between two matrices of equal size.

    Parameters:
        matrix1 (pd.DataFrame): First matrix (e.g., control or normal data).
        matrix2 (pd.DataFrame): Second matrix (e.g., experimental or tumor data).

    Returns:
        pd.DataFrame: Log2 fold change matrix.
    """
    # Ensure the matrices are of the same shape
    if matrix1.shape != matrix2.shape:
        raise ValueError("Both matrices must have the same shape.")
    
    # Add a small pseudocount to avoid division by zero
    pseudocount = 1e-6
    
    # Compute log2 fold change
    log2_fc = np.log2((matrix2 + pseudocount) / (matrix1 + pseudocount))
    
    # Return as DataFrame with same index and columns
    return pd.DataFrame(log2_fc, index=matrix1.index, columns=matrix1.columns)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Transpose to make samples rows and genes columns
data_normalized = StandardScaler().fit_transform(log_counts.T)
pca = PCA(n_components=2)  # Keep top 2 principal components
principal_components = pca.fit_transform(data_normalized)

# Convert PCA results into a DataFrame
pca_df = pd.DataFrame(data=principal_components, columns=["PC1", "PC2"], index=log_counts.columns)

print(pca_df)
```

```python
# PCA components (loadings)
loadings = pd.DataFrame(pca.components_.T, index=log_counts.loc[producing].index, columns=["PC1", "PC2"])
print(loadings.sort_values(by="PC1", ascending=True))  # Genes contributing least to PC1
```

```python
# Scatterplot of PCA
plt.figure(figsize=(8, 6))
sns.scatterplot(x="PC1", y="PC2", data=pca_df, s=100)

# Add sample names
for sample in pca_df.index:
    plt.text(pca_df.loc[sample, "PC1"] + 0.1, pca_df.loc[sample, "PC2"], sample)

plt.title("PCA of Gene Expression Data")
plt.xlabel(f"PC1 ({pca.explained_variance_ratio_[0]*100:.2f}% Variance)")
plt.ylabel(f"PC2 ({pca.explained_variance_ratio_[1]*100:.2f}% Variance)")
plt.axhline(0, color='grey', linestyle='--', linewidth=0.5)
plt.axvline(0, color='grey', linestyle='--', linewidth=0.5)
plt.grid()
plt.show()
```
